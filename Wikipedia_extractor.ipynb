```
def get_wiki(query,i):
  title= wikipedia.search(query)[i]#i is a number

  page= wikipedia.page(title)
  #Article_title=wikipedia.title(page)
  #Article_title=title.page

  return page.content
```

```
data = [] 
for i in sent_tokenize(text): 
    temp = [] 
      
    # tokenize the sentence into words 
    for j in word_tokenize(i): 
        if j.isalpha():
          j
          temp.append(j) 
  
    data.append(temp)
```
